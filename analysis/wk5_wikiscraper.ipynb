```python
# Import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Load url to scrape
headers = {'User-Agent': 'Mozilla/5.0'}
url = "https://en.wikipedia.org/wiki/List_of_countries_by_renewable_electricity_production"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Find tables
tables = soup.find('table', {'class': 'wikitable'})

# Find desired table and convert to dataframe
renew_df = pd.read_html(str(tables))[0]
renew_df
```

```python
# Filter df for SE Asia countries
renew_df = renew_df.query("Location in ['Philippines', 'Indonesia', 'Thailand', 'Vietnam', 'Malaysia']")
renew_df
```

```python
# Reshape df to long format
renew_df = renew_df.melt(
    id_vars=['Location', 'Year'],
    value_vars=['Hydro','Wind','Solar','Bio.','Geo.'],
    var_name='Source',
    value_name='TWh'
)
renew_df = renew_df.sort_values(['Location','Source'])
renew_df
```

```python
# Save df as csv
renew_df.to_csv('wk5_seasia_renew.csv', index=False)
```
