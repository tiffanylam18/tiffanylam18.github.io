```python
# Import libraries
import requests
import json
import os
```

```python
# Define API, base filename, and series to download
url_base = "https://archive-api.open-meteo.com/v1/archive?latitude={lat}&longitude={lon}&start_date=2014-01-01&end_date=2024-12-31&daily=temperature_2m_mean&timezone=auto"
file_base = "wk6_temp_{city}.json"
cities = {
    'MNL': {'lat': 14.6, 'lon': 120.98},
    'BKK': {'lat': 13.75, 'lon': 100.50},
    'JKT': {'lat': -6.21, 'lon': 106.85},
    'HCMC': {'lat': 10.82, 'lon': 106.63},
    'KL': {'lat': 3.14, 'lon': 101.69},
    'SG': {'lat': 1.29, 'lon': 103.85}
}
```

```python
# Loop to batch download:
for city, coords in cities.items():
   URL = url_base.format(lat=coords['lat'], lon=coords['lon'])
   data = requests.get(URL).json()
   filename = file_base.format(city=city)
   path = os.path.join('/Users/tiffanylam/Desktop/LSE/Year 2/PP434 Automated Data Visualization for Policymaking/tiffanylam18.github.io/data', filename)
   with open(path, 'w', encoding='utf-8') as f:
      json.dump(data, f, ensure_ascii=False, indent=4)
```
